```python
import requests
import pandas as pd
from io import StringIO
import numpy as np
import random
import math
import matplotlib.pyplot as plt


pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)

# URL of the CSV file
url = "https://data.ibb.gov.tr/dataset/a6855ce7-4092-40a5-82b5-34cf3c7e36e3/resource/56c66478-6b86-406d-a751-0ef2beb02124/download/hourly_transportation_202403.csv"

# Send an HTTP request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the content of the response as a CSV file
    csv_content = response.content.decode('utf-8')
    # Read the CSV content into a pandas DataFrame
    df = pd.read_csv(StringIO(csv_content))
    # Now you can work with the DataFrame 'df'
else:
    print("Failed to retrieve the CSV file.")

print(df)
```

##Genetic Algorithm

```python
##genetic algorithm

class GeneticAlgorithm:
    def __init__(self, population_size, mutation_rate, num_generations):
        self.population_size = population_size
        self.mutation_rate = mutation_rate
        self.num_generations = num_generations

    def initialize_population(self, num_genes):
        # Initialize the population with random schedules
        population = []
        for _ in range(self.population_size):
            schedule = np.random.randint(24, size=num_genes)
            population.append(schedule)
        return population

    def crossover(self, parent1, parent2):
        # Perform crossover between two parents
        crossover_point = np.random.randint(len(parent1))
        child1 = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
        child2 = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))
        return child1, child2

    def mutate(self, individual):
        # Perform mutation on an individual
        for i in range(len(individual)):
            if np.random.rand() < self.mutation_rate:
                individual[i] = np.random.randint(24)
        return individual

    def evaluate_fitness(self, schedule, data):
        selected_hours = data.iloc[schedule]['transition_hour']
        total_passengers = data[data['transition_hour'].isin(selected_hours)]['number_of_passenger'].sum()
        fitness = 1 / (total_passengers + 1)  # Fitness function (minimizing the total passengers)
        return fitness

    def select_parents(self, population, data):
        parent1 = population[np.random.randint(len(population))]
        parent2 = population[np.random.randint(len(population))]
        return parent1, parent2

    def evolve_population(self, population, data):
        new_population = []
        for _ in range(self.population_size):
            parent1, parent2 = self.select_parents(population, data)
            child1, child2 = self.crossover(parent1, parent2)
            child1 = self.mutate(child1)
            child2 = self.mutate(child2)
            new_population.append(child1)
            new_population.append(child2)
        return new_population

    def optimize_schedule(self, data):
        num_genes = 24
        population = self.initialize_population(num_genes)
        for _ in range(self.num_generations):
            population = self.evolve_population(population, data)

        best_schedule = max(population, key=lambda x: self.evaluate_fitness(x, data))
        return best_schedule


# Create an instance of the GeneticAlgorithm class
ga_rayli = GeneticAlgorithm(population_size=100, mutation_rate=0.01, num_generations=100)
ga_otoyol = GeneticAlgorithm(population_size=100, mutation_rate=0.01, num_generations=100)
ga_bests= []
# Loop through each road type
for road_type, df_road, ga_instance in [('RAYLI', df_rayli, ga_rayli), ('OTOYOL', df_otoyol, ga_otoyol)]:
    print(f"Optimizing schedules for {road_type}...")
    # Loop through each day of the week
    for day in range(7):
        print(f"Day {day}")
        # Filter data for the current day of the week
        df_day = df_road[df_road['day_of_week'] == day]
        # Optimize schedule for the current day using Genetic Algorithm
        best_schedule = ga_instance.optimize_schedule(df_day)
        # Get the indexes of the three lowest values in the schedule
        best_three_hours = sorted(range(len(best_schedule)), key=lambda i: best_schedule[i])[:3]
        ga_bests.append(best_three_hours)
        print(f"Best three hours for day {day}: {best_three_hours}")
```

##Particle Swarm Optimization

```python
##particle swarm optimization

# Define the Particle class
class Particle:
    def __init__(self, num_dimensions):
        self.position = np.random.randint(24, size=num_dimensions)  # Initialize position randomly
        self.velocity = np.random.rand(num_dimensions) - 0.5  # Initialize velocity randomly
        self.best_position = self.position.copy()  # Initialize personal best position as current position

# Define the ParticleSwarm class
class ParticleSwarm:
    def __init__(self, population_size, num_dimensions, num_particles, num_iterations):
        self.population_size = population_size
        self.num_dimensions = num_dimensions
        self.num_particles = num_particles
        self.num_iterations = num_iterations
        self.best_global_position = None
        self.best_global_fitness = float('inf')

    def initialize_population(self):
        return [Particle(self.num_dimensions) for _ in range(self.population_size)]

    def evaluate_fitness(self, schedule, data):
        selected_hours = data.iloc[schedule]['transition_hour']
        total_passengers = data[data['transition_hour'].isin(selected_hours)]['number_of_passenger'].sum()
        fitness = 1 / (total_passengers + 1)  # Fitness function (minimizing the total passengers)
        return fitness

    def update_global_best(self, population, data):
        for particle in population:
            fitness = self.evaluate_fitness(particle.position, data)
            if fitness < self.best_global_fitness:
                self.best_global_fitness = fitness
                self.best_global_position = np.copy(particle.position)

    def optimize_schedule(self, data):
        population = self.initialize_population()

        for _ in range(self.num_iterations):
            self.update_global_best(population, data)
            for particle in population:
                velocity = (np.random.rand(self.num_dimensions) * (particle.best_position - particle.position)) + \
                           (np.random.rand(self.num_dimensions) * (self.best_global_position - particle.position))
                particle.velocity += velocity
                particle.position = np.clip(np.round(particle.position + particle.velocity), 0, 23)

        return self.best_global_position

# Create an instance of the ParticleSwarm class for RAYLI
pso_rayli = ParticleSwarm(population_size=100, num_dimensions=24, num_particles=50, num_iterations=100)

# Create an instance of the ParticleSwarm class for OTOYOL
pso_otoyol = ParticleSwarm(population_size=100, num_dimensions=24, num_particles=50, num_iterations=100)
pso_bests = []
# Loop through each road type
for road_type, df_road, pso in [('RAYLI', df_rayli, pso_rayli), ('OTOYOL', df_otoyol, pso_otoyol)]:
    print(f"Optimizing schedules for {road_type}...")
    # Loop through each day of the week
    for day in range(7):
        print(f"Day {day}")
        # Filter data for the current day of the week
        df_day = df_road[df_road['day_of_week'] == day]
        # Optimize schedule for the current day using Particle Swarm Optimization
        best_schedule = pso.optimize_schedule(df_day)
        # Get the indexes of the three lowest values in the schedule
        best_three_hours = sorted(range(len(best_schedule)), key=lambda i: best_schedule[i])[:3]
        pso_bests.append(best_three_hours)
        print(f"Best three hours for day {day}: {best_three_hours}")

```

## Simulated Annealing

```python
## simulated annealing

class SimulatedAnnealing:
    def __init__(self, num_dimensions, num_iterations, temperature, cooling_rate):
        self.num_dimensions = num_dimensions
        self.num_iterations = num_iterations
        self.temperature = temperature
        self.cooling_rate = cooling_rate
        self.best_solution = None
        self.best_fitness = float('inf')

    def initialize_solution(self):
        return np.random.randint(24, size=self.num_dimensions)

    def evaluate_fitness(self, solution, data):
        total_passengers = data[data['transition_hour'].isin(solution)]['number_of_passenger'].sum()
        fitness = 1 / (total_passengers + 1)  # Fitness function (minimizing the total passengers)
        return fitness

    def acceptance_probability(self, new_fitness, current_fitness, temperature):
        if new_fitness < current_fitness:
            return 1
        else:
            return math.exp((current_fitness - new_fitness) / temperature)

    def anneal(self, data):
        current_solution = self.initialize_solution()
        current_fitness = self.evaluate_fitness(current_solution, data)

        for _ in range(self.num_iterations):
            new_solution = current_solution.copy()
            # Randomly select two indices to swap
            idx1, idx2 = random.sample(range(self.num_dimensions), 2)
            # Swap the values at the selected indices
            new_solution[idx1], new_solution[idx2] = new_solution[idx2], new_solution[idx1]
            new_fitness = self.evaluate_fitness(new_solution, data)
            # If the new solution is better, accept it
            if new_fitness < current_fitness or random.random() < self.acceptance_probability(new_fitness, current_fitness, self.temperature):
                current_solution = new_solution
                current_fitness = new_fitness
                if new_fitness < self.best_fitness:
                    self.best_solution = new_solution
                    self.best_fitness = new_fitness
            # Cooling down the temperature
            self.temperature *= 1 - self.cooling_rate

        return self.best_solution

temperature = 1000
cooling_rate = 0.003
num_iterations = 10000
num_dimensions = 24

# Create an instance of the SimulatedAnnealing class for RAYLI
sa_rayli = SimulatedAnnealing(num_dimensions, num_iterations, temperature, cooling_rate)

# Create an instance of the SimulatedAnnealing class for OTOYOL
sa_otoyol = SimulatedAnnealing(num_dimensions, num_iterations, temperature, cooling_rate)
sa_bests = []
# Loop through each road type
for road_type, df_road, sa in [('RAYLI', df_rayli, sa_rayli), ('OTOYOL', df_otoyol, sa_otoyol)]:
    print(f"Optimizing schedules for {road_type}...")
    # Loop through each day of the week
    for day in range(7):
        print(f"Day {day}")
        # Filter data for the current day of the week
        df_day = df_road[df_road['day_of_week'] == day]
        # Optimize schedule for the current day using Simulated Annealing
        best_schedule = sa.anneal(df_day)
        # Get the indexes of the three lowest values in the schedule
        best_three_hours = sorted(range(len(best_schedule)), key=lambda i: best_schedule[i])[:3]
        sa_bests.append(best_three_hours)
        print(f"Best three hours for day {day}: {best_three_hours}")
```
